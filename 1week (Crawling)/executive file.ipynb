```python
import time
import sys 
import os 
import pandas as pd
import numpy as np
import chromedriver_autoinstaller
from selenium.webdriver.common.by import By
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

keyword = input('')

path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(path)
driver.get(f"https://map.naver.com/v5/search/{keyword}?c=14203933.7141038,4562681.4505997,10,0,0,0,dh")
driver.switch_to.frame("searchIframe")



title_list = []
f_data_list = []

try: 
    for i in range(1,7): 
        driver.find_element(By.LINK_TEXT, str(i)).click()
        try: 
            for j in range(3,70,3):
                element = driver.find_elements(By.CSS_SELECTOR, '.place_bluelink')[j]
                ActionChains(driver).move_to_element(element).key_down(Keys.PAGE_DOWN).key_up(Keys.PAGE_DOWN).perform()
        except:
            pass

        title_raw = driver.find_elements(By.CSS_SELECTOR, ".place_bluelink")
        for title in title_raw:
            title = title.text
            title_list.append(title)

        # 평점 등 데이터
        data_raw = driver.find_elements(By.CSS_SELECTOR, '.Dr_06')
        for data in data_raw: 
            data = data.text
            f_data_list.append(data)
except:
    pass

#데이터 프레임 행, 열 길이 맞추기
count = 0
if len(title_list) != len(f_data_list):
    count = len(title_list) - len(f_data_list)
    title_list = title_list[0:len(title_list)-count]
    
print(len(title_list),len(f_data_list))
```

```python
df = pd.DataFrame({'title':title_list, 'data':f_data_list})
df
```

```python
df['data'][0:3]
```

```python
# 별점 없는 데이터 인덱스 확인
d = df[~df['data'].str.contains('별점')].index
d
```

```python
refined_df = df.drop(index=d, axis=0)
refined_df
```

```python
data_list = refined_df['data'].values.tolist()
data_list[0:5]
```

```python
for i in range(len(data_list)):
   # 별점 숫자 앞 제거
    data_list[i] = data_list[i].split('별점')[1]
    data_list[i] = data_list[i].split('\n')[1]

    # 나머지 한글 제거
    data_list[i] = data_list[i].replace('방문자리뷰', '').replace('블로그리뷰','')

    # 별점, 방문자리뷰, 저장수 분리
    data_list[i] = data_list[i].split(' ')
```

```python
data_list
```

```python
# 각 요소 리스트로 만들기

rating = []
review = []

for i in range(len(data_list)):
    rating.append(data_list[i][0])
    review.append(data_list[i][1])
```

```python
# 새로운 df 만듦
naver_df = refined_df

# 컬럼 추가 
naver_df['rating'] = rating
naver_df['review'] = review
```

```python
naver_df
```

```python
# data 컬럼 삭제 

naver_df = naver_df.drop(['data'], axis=1)
naver_df
```

```python
## csv 저장 
naver_df.to_csv("Crawling Result.csv", encoding='utf-8-sig')
```
